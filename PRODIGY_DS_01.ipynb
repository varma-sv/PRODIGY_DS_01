# Population increase from 1960 to 2023

```python
import matplotlib.pyplot as plt
import seaborn as sns

years = df.columns[1:]
years = [year for year in years if year.isdigit()]

total_population = df[years].sum()
total_population.index = total_population.index.astype(int)

total_population = total_population.sort_index()

plt.figure(figsize=(12, 10))
plt.bar(total_population.index, total_population.values, color='GREEN')
plt.xlabel('Year', size=15)
plt.ylabel('Total Population')

plt.title('Total Population over year', size=20)
plt.show()
```

# Most Populous Countries

```python
import matplotlib.pyplot as plt
import seaborn as sns
year = 1960
data = df[['Country Name', str(year)]].dropna()

data.columns = ['Country', 'Population']


data = data.sort_values(by='Population', ascending=False)

plt.figure(figsize=(30, 10))
sns.barplot(data=data.head(50), x='Country',y='Population')
plt.title(f'Top 50 Most Populous Countries in {year}')
plt.xlabel('Countries Name ')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.show()
```

# Population of particular country over year
